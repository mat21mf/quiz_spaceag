```python
#!/usr/bin/env python2

import matplotlib.pyplot as plt
import descartes
import geopandas as gpd

map_df = gpd.read_file("atributos_med.shp")
map_df.head()

vmin, vmax = map_df.med.min(), map_df.med.max()
fig, ax = plt.subplots(1, figsize=(10, 6))
map_df.plot(column=map_df.med, cmap='Blues', ax = ax, linewidth=0.8, edgecolor='0.8')
ax.axis('off')
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)
```
